```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Downloads/legend.jpg')
```

```python
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
face_cascade = cv2.CascadeClassifier('Downloads/haarcascade_frontalface_default.xml')
```

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=4, minSize=(40, 40))


for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
